```python

def spacy_annotate(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    print("Noun Phrases: ",[chunk.pos for chunk in doc])

```

```python
# Part 2 

def manual_annotate(sentence):
    full_sentence = "The cat sat on the couch"
    sentence = full_sentence.split()
    pos =      [["DET"], ["NOUN"], ["VERB"], ["PREP"], ["DET"], ["NOUN"]]

    # defining the grammar rules
    G = [
        # S is axiom
        ("S", ("NP", "VP")), 
        
            # non terminal rules
        ("NP", ("DET", "NOUN")),
        ("PP", ("PREP", "NP")),
        ("VP", ("VERB", "PP")),
        
        # terminal (if using pos directly)
        ("DET", ("DET",)),
        ("VERB", ("VERB",)),
        ("NOUN", ("NOUN",)),
        ("PREP", ("PREP",))
    ]
    cyk = CYK(G)
    print(cyk(pos, sentence)) 
```
